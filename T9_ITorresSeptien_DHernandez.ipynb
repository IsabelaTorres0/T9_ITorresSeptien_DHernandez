## Subamortiguado

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import odeint

k, m, c = 3, 1, 0.5

def amortiguado(y,t,k,m,c):
    y1 = y[0]
    y2 = y[1]
    return np.array([y2, -(k/m)*y1-(c/m)*y2])

coni0 = np.array([1,1])
coni1 = np.array([1,-1])
coni2 = np.array([-1,1])
coni3 = np.array([-1,-1])

t = np.linspace(0, 30, 300)

y = odeint(amortiguado, coni0, t, args=(k,m,c))
y1su = y[:,0]
y = odeint(amortiguado, coni1, t, args=(k,m,c))
y2su = y[:,0]
y = odeint(amortiguado, coni2, t, args=(k,m,c))
y3su = y[:,0]
y = odeint(amortiguado, coni3, t, args=(k,m,c))
y4su = y[:,0]
plt.figure(figsize=(8,6))
plt.plot(t, y1su, 'b', lw=3, label='Posición [m]: $y_1(t)$')
plt.plot(t, y2su, 'r', lw=3, label='Posición [m]: $y_2(t)$')
plt.plot(t, y3su, 'g', lw=3, label='Posición [m]: $y_3(t)$')
plt.plot(t, y4su, 'y', lw=3, label='Posición [m]: $y_4(t)$')
plt.xlabel("Tiempo [s] $t$")
plt.legend(loc="best")
plt.grid()
plt.show()
```

# Conclusión estudiante 1:

¿Cómo se ve el espacio fase para los diferentes casos así como para diferentes condiciones iniciales?

Como se puede observar de manera clara en las 12 graficas que se presentan en la parte superior, el movimiento subamortiguado permite que se realicen unas pocas oscilaciones antes de realmente verse afectada la velocidad de modo que el movimiento será lentamente disminuido.
Por su parte las que están sobre amortiguadas, se puede observar que el tiempo que le toma para llegar al "reposo" es muy bajo, es muy dificil para el sistema moverse por lo que para los que solo tienen una condicion incial negativa se llega bastante rapido al equilibrio mientras que a los que tienen ambas del mismo signo les cuesta un poco alcanzarlo ya que incluso de ese modo le cuesta regresar a la posición original.
Con el crítico por su parte se puede observar que parece llegar más rapido al equilibrio cuando ambas condiciones iniciales son positivas pero es más dificil alcanzarlo cuando una de ellas es negativa ya que este permite un pequeño "rebote" al llegar al lado opuesto del origen.

Código de solución estudiante 2

**SUBAMORTIGUADO**

```python
# Primero importamos las librerias,
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import odeint
```

```python
# Definimos los parámetros k, m y c
k, m, c = 3, 1, 0.5

# Función f(y,t) que vamos a integrar
def amortiguado(y,t,k,m,c):
    y1 = y[0]
    y2 = y[1]
    return np.array([y2, -(k/m)*y1-(c/m)*y2])

# Condiciones iniciales
c0 = np.array([1,1])
c1 = np.array([1,-1])
c2 = np.array([-1,1])
c3 = np.array([-1,-1])

# Especificamos los puntos de tiempo donde queremos la solución
t = np.linspace(0, 30, 300)

# Solución numérica
y = odeint(amortiguado, c0, t, args=(k,m,c))
sub0 = y[:,0]
y = odeint(amortiguado, c1, t, args=(k,m,c))
sub1 = y[:,0]
y = odeint(amortiguado, c2, t, args=(k,m,c))
sub2 = y[:,0]
y = odeint(amortiguado, c3, t, args=(k,m,c))
sub3 = y[:,0]
```

```python
# Gráfica
plt.figure(figsize=(8,6))
plt.plot(t, sub0, 'b', lw=3, label='Posición [m]: $y_1(t)$')
plt.plot(t, sub1, 'r', lw=3, label='Posición [m]: $y_2(t)$')
plt.plot(t, sub2, 'p', lw=3, label='Posición [m]: $y_3(t)$')
plt.plot(t, sub3, 'x', lw=3, label='Posición [m]: $y_4(t)$')
plt.xlabel("Tiempo [s] $t$")
plt.legend(loc="best")
plt.grid()
plt.show()
```
